# Distribution Plots

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\Eddie\\Downloads\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\dm_office_sales.csv')
df.head()
```

```python
#plt.figure(figsize=(5,8))
sns.rugplot(x = 'salary', data = df, height= 0.5)
```

```python
sns.set(style = 'ticks')
sns.displot(x = 'salary', data = df, bins = 20, color = 'red', edgecolor = 'blue', linewidth = 4, kde = True, rug = True)
```

```python
sns.histplot(x = 'salary', data = df)
```

```python
sns.kdeplot(data = df, x = 'salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)  # give me 200 numbers between 0-100

sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns = ['age'])
```

```python
sample_ages
sns.rugplot(x = 'age', data = sample_ages)
```

```python
sns.displot(x = 'age', data = sample_ages, rug = True, bins = 30, kde = True)
```

```python
sns.kdeplot(x= 'age',data = sample_ages, clip = [0,100], bw_adjust = 0.5, shade = True)
# does give some probablity outside of data range unless use clip
# bw_adjust allows you to play with the level of signal the graph picks up on
# whether you want it to be more or less sensitive to data
```
