### Two main goals
* be able to plot out functional relation ships e.g y = 2x
* be able to plot out a relationship between two raw data points

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10, 0.1)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(0,6)              # refines the end and starting point for the x axis
                           # otherwise it will show a little before your first data point and a little after your last
plt.ylim(-5,5)
plt.title('String Title');  # semicolon gets rid of useless text output
plt.savefig('myfirstplot.png')
```

# Figure Object

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])    # 0,0 represent the lower left hand corner of the plot 
                                  # starting position of x,y
                                  # if you started at 0,1 you would start in the top left hand corner of the plot 
                                  # width and height (1,1) are with respect to the figure size
                                  # e.g the figure is 100x100 pixels then the (0.5,0.5) means the axes extends 50x50 pixels
                                  # the x and y axes will still go from 0-1
axes.plot(x,y)
```

# Figure Objects
### Implementing Figures and Axes
                                  
        


```python
# Data 
a = np.linspace(0,10,11)   # 11 points between 0,10
b = a ** 4
```

```python
x =np.arange(0,10)         # 0-9 incrementing by 1
y = 2*x
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b);
```

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

# SMALL AXES
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(a,b);
axes2.set_xlim(1,2)
axes2.set_ylim(0,20)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in')
```

# Figure Parameters

```python
fig = plt.figure(figsize = (2,2), dpi = 200)  
# dpi makes the figure significantly less blurry / tradeoff takes more ram to load high dpi figure
# fig size allows you to enlarge or reduce figure size 
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig('new_figure.png', bbox_inches = 'tight')
```

# Subplots
#### plt.subplots() returns a tuple which by common convention we label (fig, axes):
* fig - this is the entire figure canvas
* axes - this is a numpy array holding each of the axes according to position in the overall canvas 

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)

plt.tight_layout()
# automatically adjusts plots so that no values are overlapping
```

### Manual formatting of subplots

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)

fig.subplots_adjust(wspace = 1, hspace = 0.5)
#wspace = width space between figures 
#hspace = height space between figures
# values of spaces are fractions - fractions of average figure size
# e.g. a value of 1 means 1 whole figure could fit between the plots

axes[0][0].set_title("1st")
axes[0][0].set_ylabel('Y label')
axes[0][0].set_xlim(0,6)
# for setting titles/labels/etc.. it is normal commands just have to specify which set of axes you're referring to

fig.suptitle("FIGURE LEVEL", fontsize = 16)
# Figure level titles

fig.savefig('subplots.png' , bbox_inches = 'tight')
```

# Figure Styling - Legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label = 'X vs X')
ax.plot(x,x**2, label = 'X vs X^2')
ax.legend(loc = (1.1,0.5))
# can add parameter loc = "lower left or upper right or ..."
# also can pass in a tuple
```

# Figure Styling - Colors and Styles

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color = '#e6c017', label = 'X vs X')
# color can take named values like 'purple'
# color can also take RGB hex code

ax.plot(x,x+1, color = 'purple', label = 'X vs X+1', lw = 4, ls = '-.')
# lw -> line width default value is 1 ( can also type out linewidth fully )
# ls -> line style can take in different values like '- , -- , : , -. '


ax.plot(x,x-1, color = 'red', label = 'X vs X-1', lw =2, marker = '^', markersize = 8, markerfacecolor = 'red',
        markeredgewidth = 2, markeredgecolor = 'orange')
# ms -> marker size

ax.legend(loc = 'upper left')
```
