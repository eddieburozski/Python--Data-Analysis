```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\Eddie\\Downloads\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\StudentsPerformance.csv")
df.head()
```

```python
sns.catplot(data = df, x = 'gender', y = 'math score', kind = 'box', col = 'lunch', row = 'test preparation course')
plt.show()

# row = '...' gives separate subplot for each category of the named column
# col does same as row but obviously presents as a column
# similar to hue but hue gives all information in one plot
```

```python
sns.catplot(data = df, x = 'gender', y = 'math score', kind = 'violin', col = 'lunch', row = 'race/ethnicity')
plt.show()
```

```python
g = sns.PairGrid(data = df, hue = 'gender')
# PairGrid is very similar to pairplot however pair grid allows for control over grids on the axes, as well as above and below
# the axes
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.kdeplot)

g = g.add_legend()
plt.show()
```
