# Scatterplots in Seaborn


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\Eddie\\Downloads\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize = (12,4), dpi = 200)
sns.scatterplot(x = 'salary',y = 'sales',data = df, hue = 'level of education', palette = 'Dark2')

# hue takes a column name and colors the data by that column
# if it is categorical, like example, then it breaks it up nice and easy
# if it is continuous it will break it up into arbitrary ranges - making a gradient

# seaborn is calling matplotlib on the backend
# so if prior to your seaborn call you start editing a plt.figure seaborn is smart enough to use that figure
```

```python
plt.figure(figsize = (12,4), dpi = 200)
sns.scatterplot(x = 'salary',y = 'sales',data = df, size = 'work experience', alpha = 0.2, style = 'level of education')

# sizing works best for continuous data
# s = 200 makes all points the same size of 200

# for many overlapping points use alpha to show overlap - transparency 

plt.savefig('my_plot.jpg')
```
