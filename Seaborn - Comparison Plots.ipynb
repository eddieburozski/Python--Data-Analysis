```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\Eddie\\Downloads\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\StudentsPerformance.csv")
```

```python
# we want to find if someone scores well in one subject are they likey to score well in other subjects as well?
sns.jointplot(data = df, x = 'math score', y = 'reading score', kind = 'kde', shade = True)

# kind = 'hex' changes the scatter plot from lots of overlapping dots to hexagons that are shaded with respect to density
# kind = 'hist' returns similar to above but with squares also they line up with the bins of the histogram
# kind = 'kde' returns density mapping as well but it is an outline unless add shade = True
```

```python
sns.jointplot(data = df, x = 'math score', y = 'reading score',hue = 'gender')
```

```python
# pairplot is CPU intensive so be careful when calling it
# grabs all numerical columns and compares them 
sns.pairplot(data = df, hue = 'gender', corner = True)

# corner gets rid of duplicates
```
