## A groupby() operation allows us to examine data on a per category basis


```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Eddie\\Downloads\\UNZIP_FOR_NOTEBOOKS_FINAL\\03-Pandas\\mpg.csv")
```

```python
df['model_year'].value_counts()
```

```python
df.groupby('model_year').mean()   # returns the averages of all the numeric columns grouped by the specified 'model_year'
```

```python
df.groupby('model_year').mean()['mpg']
```

```python
# but what if you want to group by mulitple categories? 

df.groupby(['model_year','cylinders']).mean()
# df.groupby(['model_year','cylinders']).mean().columns      returns a list of the names of the columns
```

```python
year_cyl = df.groupby(['model_year','cylinders']).mean()
```

```python
year_cyl
```

```python
year_cyl.index.levels
```

```python
year_cyl.loc[[70,82]]
```

```python
# if you want to select a year and cylinder type must pass in a tuple
year_cyl.loc[[(70,4),(80,6)]]
```

```python
year_cyl.xs(key = 70, level = 'model_year')
```

```python
four_cyl = year_cyl.xs( key = 4, level = 'cylinders')
```

```python
# what if you want to look at cars with 6 cylinders AND 8 cylinders
# xs key can only take one value therefore you want to refine/filter your data before this point
df[df['cylinders'].isin([6,8])].groupby(['model_year','cylinders']).mean()
```

```python
# what if you want to swap levels?
year_cyl.swaplevel()
```

```python
# Sorting!
year_cyl.sort_index(level = 'model_year', ascending = False)
```

```python
df.agg(['std','mean'])['mpg']
```

```python
df.agg({'mpg':['max','mean'], 'weight':['mean','std']})    # more advanced way to perform aggregations on specific columns
```
