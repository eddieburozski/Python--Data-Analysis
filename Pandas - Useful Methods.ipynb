## Useful Methods

* will cover a wide range of method calls in Pandas

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:\\Users\\Eddie\\Downloads\\UNZIP_FOR_NOTEBOOKS_FINAL\\03-Pandas\\tips.csv')
```

```python
df.head()
```

```python
# .apply() allows you to add any custom function to every row in a Series
def last_four(num):
    '''
    Function returns last four digits of credit card as a string
    '''
    return str(num)[-4:]
```

```python
last_four(123456)
```

```python
df['last four'] = df['CC Number'].apply(last_four)     # notice you're not calling the function e.g. last_four() 
                                                       # pandas calls the function for each element of the column
```

```python
df.head()
```

```python
df['total_bill'].mean()
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
    
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```

```python
df.head()
```

## Applying Functions to Multiple Columns

```python
def simple(num):
    return num*2
# convert to lambda
lambda num: num*2
```

```python
df['total_bill'].apply(simple)
# EQUIVALENT EXPRESSIONS
df['total_bill'].apply(lambda num: num*2)
```

```python
df.head()
```

```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return 'Other'
```

```python
quality(16.99, 1.01)
```

```python
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'],df['tip']), axis = 1)
```

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])    # Equivalent expressions 
                                                                     # vectorize is more nice to type also it typically 
                                                                     # performs better
```

## Replacing Data in your Dataset

```python
df['sex'].replace("Female","F")

df['sex'].replace(["Female","Male"], ['F','M'])   # Can also take multiple values, second list just has to be in same order
```

```python
# Another way to replace but with dictionaries

mymap = {"Female":"F", "Male":"M"}                     # Better syntax with this approach
df['sex'].map(mymap)
```
